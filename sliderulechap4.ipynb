```python
mydata = pd.DataFrame.from_csv('./WDI_Data.csv', index_col=None,
                               encoding='utf-8')
#mydata = mydata.drop('CCode',1)
mydata = mydata.drop(mydata.columns[[1,3]],1)
mydata.head()
```

```python
m = mydata.set_index(['Country','IndName'])
m.head()
```

```python
mplt = m.loc[('India','Access to electricity (% of population)')]

mplt = mplt.fillna(0)

mpltus = m.loc[('United States','Access to electricity (% of population)')]
mpltus = mpltus.fillna(0)

mpltch = m.loc[('China','Access to electricity (% of population)')]
mpltch = mpltch.fillna(0)

plt.figure()
mplt.plot(ylim=[0,100],  style='*',color='r',label = 'India')
mpltus.plot(ylim=[0,100],style='^',color='g',label='USA')
mpltch.plot(ylim=[0,100],style='+',color='m',label='China')

plt.xlabel('year')
plt.ylabel('Access to electricity (% of population)')
plt.title('Electricity')
plt.legend()
```

```python
mplt = m.loc[('India','Adjusted net national income per capita (annual % growth)')]

mplt = mplt.fillna(0)

mpltus = m.loc[('United States','Adjusted net national income per capita (annual % growth)')]
mpltus = mpltus.fillna(0)

mpltch = m.loc[('China','Adjusted net national income per capita (annual % growth)')]
mpltch = mpltch.fillna(0)

plt.figure()
mplt.plot(ylim=[0,40],  style='-',color='r',label = 'India')
mpltus.plot(ylim=[0,40],style='-',color='g',label='USA')
mpltch.plot(ylim=[0,40],style='-',color='m',label='China')

plt.xlabel('year')
plt.ylabel('Adjusted net national income per capita (annual % growth)')
plt.title('Annual Income Per Capita ')
plt.legend()
```
